```python
def sum_range(start, end):
  if start > end:
    end, start = start, end
  return sum(range(start, end + 1))

print(sum_range(2, 12))
print(sum_range(-4, 4))
print(sum_range(3, 2))
```

## 1.2.3
### –ü—Ä–∏–º–µ—Ä
> –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ Python, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ
–º–µ–∂–¥—É –¥–≤—É–º—è –º–∞—Å—Å–∏–≤–∞–º–∏ NumPy.

```python
import numpy as np
def euclidean_distance(v1, v2):

  return sum((x - y) ** 2 for x, y in zip(v1, v2)) ** 0.5

x = np.array([0,0,0])
y = np.array ([3, 3, 3])
print(euclidean_distance(x, y))
```

## 1.2.4
### –ü—Ä–∏–º–µ—Ä
> –ù–∞–ø–∏—à–∏—Ç–µ 4 —Ñ—É–Ω–∫—Ü–∏–∏ –≤ Python, –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—Ç –∫–≤–∞–¥—Ä–∞—Ç –ï–≤–∫–ª–∏–¥–æ–≤–∞
—Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è, –≤–∑–≤–µ—à–µ–Ω–Ω–æ–µ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –•–µ–º–º–∏–Ω–≥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –∏
—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ß–µ–±—ã—à–µ–≤–∞ –º–µ–∂–¥—É –¥–≤—É–º—è –º–∞—Å—Å–∏–≤–∞–º–∏ NumPy.

```python
def sqr_euclidean_distance(v1, v2):
  return sum((x - y) ** 2 for x, y in zip(v1, v2))
def weighted_euclidean_distance(v1, v2, w):
  return sum((x - y) ** 2 * s for x, y, s in zip(v1, v2, w)) ** 0.5
def manhattan_distance(v1, v2):
  return sum(abs(x - y) for (x, y) in zip(v1, v2))
def chebyshev_distance(v1, v2):
  return max(abs(x - y) for (x, y) in zip(v1, v2))
  

x = np.array([0,0,0])
y = np.array([3,3,3])
w = np.array([0,0,1])
print(sqr_euclidean_distance(x, y))
print(weighted_euclidean_distance(x, y, w))
print(manhattan_distance(x, y))
print(chebyshev_distance(x,y))
```

## 2.2.1
### –ü—Ä–∏–º–µ—Ä

> –í –ø—Ä–∏–º–µ—Ä–µ –ø–æ–∫–∞–∑–∞–Ω–æ —Å–æ–∑–¥–∞–Ω–∏–µ 2d-–º–∞—Å—Å–∏–≤–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ x –∏ y. –°–ø–∏—Å–æ–∫
target —Å–æ–¥–µ—Ä–∂–∏—Ç –≤–æ–∑–º–æ–∂–Ω—ã–µ –≤—ã—Ö–æ–¥–Ω—ã–µ –∫–ª–∞—Å—Å—ã (—á–∞—Å—Ç–æ –Ω–∞–∑—ã–≤–∞–µ–º—ã–µ
–º–µ—Ç–∫–∞–º–∏). –î–∞–ª–µ–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –æ–±—É—á–µ–Ω–µ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ k-–±–ª–∏–∂–∞–π—à–∏—Ö
—Å–æ—Å–µ–¥–µ–π –ø–æ –∏—Å—Ö–æ–¥–Ω—ã–º –¥–∞–Ω–Ω—ã–º. –î–∞–ª–µ–µ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—Å—è –ø—Ä–æ–≥–Ω–æ–∑
–ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –∫–ª–∞—Å—Å–∞–º –¥–ª—è –¥–≤—É—Ö —Ç–æ—á–µ–∫ –¥–∞–Ω–Ω—ã—Ö.

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# –¥–∞–Ω–Ω—ã–µ
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0,0,0,1,1,1]

# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å k-–±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –∫ –¥–∞–Ω–Ω—ã–º
K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X, target)
print(model)

#  –¥–µ–ª–∞–µ–º –ø—Ä–æ–≥–Ω–æ–∑
print('(-2, =2) is class', model.predict([[-2, -2]]))

print('(1, 3) is class', model.predict([[1, 3]]))
```

## 2.2.4
> –ò–∑ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤–∏–¥–Ω–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤–æ–º —Å–ª—É—á–∞–µ –∫–ª–∞—Å—Å—ã –≤–∏–∑—É–∞–ª—å–Ω–æ —Ö–æ—Ä–æ—à–æ –æ—Ç–¥–µ–ª–∏–º—ã –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞, —Ö–æ—Ç—è –¥–≤–∞ –∫–ª–∞—Å—Å–µ –∏–º–µ—é—Ç –Ω–µ–±–æ–ª—å—à–æ–µ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ. –í–æ –≤—Ç–æ—Ä–æ–º —Å–ª—É—á–∞–µ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –¥–≤–∞ –∫–ª–∞—Å—Å–∞ –º–µ–∂–¥—É —Å–æ–±–æ–π —É–∂–µ –Ω–∞–º–Ω–æ–≥–æ —Ç—Ä—É–¥–Ω–µ–µ.
–î–∞–ª–µ–µ —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 80:20. –û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞ (training sample) ‚Äî –≤—ã–±–æ—Ä–∫–∞, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—Å—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ (–æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤) –º–æ–¥–µ–ª–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏. –¢–µ—Å—Ç–æ–≤–∞—è (–∏–ª–∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–∞—è) –≤—ã–±–æ—Ä–∫–∞ (test sample) ‚Äî –≤—ã–±–æ—Ä–∫–∞, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:,:-1],
    iris.iloc[:, -1], #iloc –≤—ã–±–∏—Ä–∞–µ—Ç —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–ª–±—Ü—ã –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –ø–æ–∑–∏—Ü–∏—è—Ö
    test_size = 0.20
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
X_train.head()
y_train.head()

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=X_test, hue=y_pred, s=70)
plt.xlabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.ylabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.legend()
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='r', s=150)
print(f'accuracy{accuracy_score(y_test, y_pred) : .3}')
```

–£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –¥—Ä—É–≥–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π (k = 1, 5, 10).
–£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ 15% –æ—Ç –≤—Å–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞.

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:,:-1],
    iris.iloc[:, -1],
    test_size = 0.15
)

model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=X_test, hue=y_pred, s=70)
plt.xlabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.ylabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.legend()
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='r', s=150)
print(f'accuracy k=1: {accuracy_score(y_test, y_pred) : .3}')

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=X_test, hue=y_pred, s=70)
plt.xlabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.ylabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.legend()
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='r', s=150)
print(f'accuracy k=5: {accuracy_score(y_test, y_pred) : .3}')

model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=X_test, hue=y_pred, s=70)
plt.xlabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.ylabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞, —Å–º')
plt.legend()
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='r', s=150)
print(f'accuracy k=10: {accuracy_score(y_test, y_pred) : .3}')
```

```python
import pandas as pd
#—Å–æ–∑–¥–∞—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏
dataframe = pd.DataFrame({"–û—Ü–µ–Ω–∫–∞": ['–Ω–∏–∑–∫–∞—è', '–Ω–∏–∑–∫–∞—è', '—Å—Ä–µ–¥–Ω—è—è', '—Å—Ä–µ–¥–Ω—è—è', '–≤—ã—Å–æ–∫–∞—è']})
#—Å–ª–æ–≤–∞—Ä—å –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
scale_mapper = {"–Ω–∏–∑–∫–∞—è": 1, "—Å—Ä–µ–¥–Ω—è—è": 2, "–≤—ã—Å–æ–∫–∞—è": 3}
#–∑–∞–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Å–ª–æ–≤–∞—Ä—è
dataframe["–û—Ü–µ–Ω–∫–∞"].replace(scale_mapper)
```
