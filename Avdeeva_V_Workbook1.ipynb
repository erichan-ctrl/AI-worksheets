## 4.2.2
> Для функции sin⁡(xe^cos⁡(x) ) на интервале [0;10] с шагом 0,05 построить график функции, закрасить площадь под ним и вычислить значение этой площади.
Для вычисления площади воспользуйтесь функцией trapz(), вычисляющей интеграл по правилу трапеции. Для ее корректной работы необходимо подключить следующие библиотеки:
 ``` python

```


````python
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz

x = np.arange(0.0, 11, 0.5)
y = np.abs(np.sin(x*np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c = 'r') #рисует точки соединенные линиями
plt.fill_between(x, y)

area = trapz(y)
print(area)
````

## 4.3.2
> Дана функция.  Создать массив из 10 значений функции (x, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива, используя plt.plot(), для среза – точечный (plt.scatter()).

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11, 1)
x1 = x[0:x.size // 2] #срез первой половины массива
y = np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(np.power(x, 2))) / np.abs(1 - np.power(np.sin(x), 3)) + np.log(np.abs(2 * x))
y1 = np.sqrt(1 + np.exp(np.sqrt(x1)) + np.cos(np.power(x1, 2))) / np.abs(1 - np.power(np.sin(x1), 3)) + np.log(np.abs(2 * x1))
plt.grid()
plt.plot(x, y, c = 'r')
plt.show()
plt.grid()
plt.scatter(x1, y1, c = 'r')
plt.show()
````
